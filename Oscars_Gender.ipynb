```python
#This code is adapted from Stackoverflow (https://stackoverflow.com/questions/42006346/pandas-convert-columns-to-percentages-of-the-totals)
#The data is made into percentages to control for variations in datasets and to show propotional variations of male and women in films
cols = ['Male Speech', 'Female Speech']
dfP = df[cols].div(df[cols].sum(axis=1), axis=0).multiply(100)
dfP
```

```python
#Code for visualisation adapted from Stackoverflow (https://stackoverflow.com/questions/27694221/using-python-libraries-to-plot-two-horizontal-bar-charts-sharing-same-y-axis)

import matplotlib.pyplot as plt

y = df.iloc[:,0]
x1 = dfP.iloc[:,0]
x2 = dfP.iloc[:,1]

fig, axes = plt.subplots(ncols=2, sharey=True)
axes[0].barh(y, x1, align='center', color='lightblue')
axes[0].set_xlim(0, 100)
axes[0].set(title='Male Characters Speaking')
axes[1].barh(y, x2, align='center', color='pink')
axes[1].set_xlim(0, 100)
axes[1].set(title='Female Characters Speaking')
axes[0].invert_xaxis()
plt.show()
```
