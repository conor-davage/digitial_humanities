```python
#For our group project we employ scraping methods to collect our data from the BBC and RTE 
#We scraped the most recent articles on the topic of Brexit from the countries' national broadcasters
#Our sample size is 24 articles as to have enough data but not to overload the webpages for ethical reasons

import requests as re
```

```python
from bs4 import BeautifulSoup as bs
```

```python
urlBBC = ('https://www.bbc.co.uk/news/uk-politics-47206286',
                'https://www.bbc.co.uk/news/uk-politics-47206342',
                'https://www.bbc.co.uk/news/business-47206303',
                'https://www.bbc.co.uk/news/uk-northern-ireland-47201546'
                'https://www.bbc.co.uk/news/uk-politics-47192233',
                'https://www.bbc.co.uk/news/uk-47187491',
                'https://www.bbc.co.uk/news/uk-politics-47170369',
                'https://www.bbc.co.uk/news/uk-politics-47152035',
                'https://www.bbc.co.uk/news/uk-politics-47143135',
                'https://www.bbc.co.uk/news/uk-northern-ireland-47110707',
                'https://www.bbc.co.uk/news/uk-northern-ireland-politics-44615404'
                'https://www.bbc.co.uk/news/uk-politics-47219386',
               )

urlRTE = ('https://www.rte.ie/news/business/2019/0213/1029287-philip-lane-at-european-financial-forum/',
               'https://www.rte.ie/news/brexit/2019/0213/1029249-may-mps-brexit/',
               'https://www.rte.ie/news/brexit/2019/0213/1029315-brexit-politics/',
               'https://www.rte.ie/eile/brainstorm/2019/0208/1029032-heres-the-moral-case-for-a-second-brexit-referendum/',
                'https://www.rte.ie/news/2019/0212/1029016-may_brexit/',
                'https://www.rte.ie/news/analysis-and-comment/2019/0212/1029004-brexit-westminster/',
                'https://www.rte.ie/news/brexit/2019/0211/1028784-brexit-theresa-may/',
                'https://www.rte.ie/news/uk/2019/0212/1029101-michael-d-higgins-england-visit/',
                'https://www.rte.ie/news/brexit/2019/0211/1028894-brexit-ireland/',
                'https://www.rte.ie/news/brexit/2019/0210/1028639-blair-brexit-ireland/',
                'https://www.rte.ie/news/brexit/2019/0208/1028244-brexit/',
                'https://www.rte.ie/news/brexit/2019/0208/1028467-brexit-tony-connelly/'
               )
```

```python
sourcesBBC = []

for page in urlBBC:
    url = re.get(page)
    print(url)
    content = url.content
    print(content)
    pages = bs(content, 'html.parser')
    print(pages.title)
    for i in pages.find_all('p')[12:]:
        text = i.text
        print(text)
        sourcesBBC.append(text)
```

```python
sourcesBBC
```

```python
sourcesRTE = []

for page in urlRTE:
    url = re.get(page)
    print(url)
    content = url.content
    print(content)
    pages = bs(content, 'html.parser')
    print(pages.title)
    for i in pages.find_all('p')[3:]:
        text = i.text
        print(text)
        sourcesRTE.append(text)
```

```python
sourcesRTE
```

```python
file = open('BBCtext.txt', 'w', encoding='utf-8')
file.write(''.join(sourcesBBC))
file.close()  
```

```python
file = open('RTEtext.txt', 'w', encoding='utf-8')
file.write(''.join(sourcesRTE))
file.close()  
```

```python
#Once we have the textual data, we then begin the process of cleaning it to ensure it is in the correct format to work with


BBCplain = BBCRead.replace('.', '').replace(',', '').replace('?', '').replace('&', '').replace('-','').replace('','').replace('&apos', '').replace('"', '').lower()
BBCplain
```

```python
RTEplain = RTERead.replace('.', '').replace(',', '').replace('?', '').replace('&', '').replace('-','').replace('','').replace('&apos', '').replace('"', '').replace('@', '').lower()
RTEplain
```

```python
BBCList = BBCplain.split()
print(BBCList)
```

```python
RTEList = RTEplain.split()
print(RTEList)
```

```python
import nltk
nltk.download()
```

```python
#Next we import stopwords and remove them from the dataset to further clean it

from nltk.corpus import stopwords

stopWords = stopwords.words('english')
   
BBCClean = []

for w in BBCList:
    if w not in stopWords:
        BBCClean.append(w)
        
print(BBCClean)
```

```python
stopWords = stopwords.words('english')
   
RTEClean = []

for w in RTEList:
    if w not in stopWords:
        RTEClean.append(w)
        
print(RTEClean)
```

```python
#Once the data is cleaned, we begin to count the frequencies of the words
#This is just an initial exploration of the data to understand the most relevant topics

from collections import Counter
BBCCount = Counter(BBCClean)
BBCCount
```

```python
from collections import Counter
RTECount = Counter(RTEClean)
RTECount
```

```python
BBCDict = {}

for word in BBCClean:
    BBCDict[word] = BBCClean.count(word)

print(BBCDict)
```

```python
RTEDict = {}

for word in RTEClean:
    RTEDict[word] = RTEClean.count(word)

print(RTEDict)
```

```python
#The data was exported from excel in csv format and imported accordingly

media_data = pd.read_csv('mediafreqdata.csv')
media_data
```

```python
#We then created variables and assigned them values of their corriesponding column
#This was in preparation for visualising the data

words = media_data.iloc[:,1]
BBC_words = media_data.iloc[:,2]
RTE_words = media_data.iloc[:,3]

indx = np.arange(len(media_data))
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#By using a stacked bar chart to visualise the data we can compare both datasets at the same time quite easily
#It illustrates both how often topics are reported on, and also how reporting is similar and different between RTE and BBC

def stackBarChart():
    plt.figure(figsize=(20, 12))
    graphBBC = plt.bar(x=indx, height=BBC_words, width=0.35)
    graphRTE = plt.bar(x=indx, height=RTE_words, width=0.35, bottom=BBC_words)
    plt.xlabel('Words', fontsize=20, color='black')
    plt.ylabel('Frequency', fontsize=20, color='black')
    plt.xticks(indx, words)
    plt.legend((graphBBC[0], graphRTE[0]), ('BBC', 'RTE'))
    plt.title('Frequency of Words in BBC vs RTE Reporing of Brexit')

stackBarChart()
```

```python
#However, this information, although interesting, is somewhat flat without knowing the context in which words are used
#In order to use the concordance method of nltk for our own text, we must tokenise the words first as below

t = open('bbctext.txt')
bbc_text = t.read()
from nltk import word_tokenize
tokenized_bbc_text = nltk.word_tokenize(bbc_text)
bbc_text = nltk.Text(tokenized_bbc_text)
```

```python
t = open('rtetext.txt')
rte_text = t.read()
from nltk import word_tokenize
tokenized_rte_text = nltk.word_tokenize(rte_text)
rte_text = nltk.Text(tokenized_rte_text)
```

```python
#We then look for the context of the words 'referendum', 'backstop', 'deal', and 'border' to indicate variations of reporting
#This shows some interesting inights and highlights a more negative angle by the Irish press on some topics
#For example, RTE uses words in context of 'second referendum', 'no deal' and more interest in the 'hard border'

bbc_text.concordance('referendum')
```
